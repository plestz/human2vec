```python
import numpy as np
import pandas as pd

import igraph as ig
from igraph import Graph

import regex as re

import os
```

```python
# -- Extract all existing Ego IDs -- 

data_path = 'data/ego-networks/facebook/'

pattern = re.compile(r'\d+')

EGO_IDS = list()

for file_path in os.listdir(data_path):

    if not file_path.endswith('.edges'):
        continue

    match_obj = pattern.match(file_path)

    if match_obj:
        EGO_IDS.append(match_obj.group())

EGO_IDS = list(map(int, EGO_IDS))
```

```python
# -- Turn raw Ego edges into Pandas DataFrame -- 

edge_df = pd.read_csv(f'data/ego-networks/facebook/{EGO_IDS[5]}.edges', sep = ' ', names = ["Node A", "Node B"])

print(edge_df)

undirected_adjacency_list = list(zip(edge_df['Node A'], edge_df['Node B']))

print(undirected_adjacency_list)
```

```python
# -- Turn Pandas DataFrame into Undirected igraph Graph -- 

# Note: While the FB dataset is undirected, its .edges file contains directed edges. In other words, it contains (A -> B) and (B -> A). 
g = Graph(edges = undirected_adjacency_list, directed = True)

node_neighbors = g.to_list_dict() # Maps Node ID X to all Neighbor IDs [V_1, ..., V_N]
print(node_neighbors)

layout = g.layout('kk') # Kamada-Kawai

visual_style = {
    'layout': layout,
    'vertex_size': 10,
}

ig.plot(g, **visual_style)
```
